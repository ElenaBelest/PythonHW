**Задача 1:** На складе лежат разные фрукты в разном количестве.
Нужно написать функцию, которая на вход принимает любое количество названий фруктов и их количество, а возвращает общее количество фруктов на складе

**Задача 2:** Дан список с затратами на рекламу. Но в данных есть ошибки, некоторые затраты имеют отрицательную величину. Удалите такие значения из списка и посчитайте суммарные затраты
[100, 125, -90, 345, 655, -1, 0, 200]
Используйте list comprehensions

```python
ztr_origin=[100, 125, -90, 345, 655, -1, 0, 200] 
ztr = [i if i>0 else 0 for i in ztr_origin]
ztr
```

```python
all_ztr = sum(ztr)
print(f'Сумма затрат на рекламу: {all_ztr}')
```
